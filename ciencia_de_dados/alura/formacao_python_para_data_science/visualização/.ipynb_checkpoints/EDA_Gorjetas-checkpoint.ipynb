# Importando Base de Dados

```python
import pandas as pd
```

## Teste de hipótese

**H<sup>null</sup>**

**A distribuição da taxa da gorjeta é a mesma nos dois grupos**

**H<sup>alt</sup>**

**A distribuição da taxa da gorjeta não é a mesma nos dois grupos**

```python
from scipy.stats import ranksums
```

```python
sobremesa = gorjetas.query("sobremesa == 'Sim'").porcentagem
```

```python
sem_sobremesa = gorjetas.query("sobremesa == 'Não'").porcentagem
```

```python
r = ranksums(sobremesa, sem_sobremesa)
```

```python
print("p-value = {}".format(r.pvalue.round(3)))
```

**Como P-value > 0.05, então, aceita-se H<sup>null<sup/>:** A distribuição da taxa da gorjeta é a mesma nos dois grupos

## Teste de hipótese

**H<sup>null</sup>**

**A distribuição do valor da conta é igual no Sábado e no Domingo**

**H<sup>alt</sup>**

**A distribuição do valor da conta não é igual no Sábado e no Domingo**

```python
valor_conta_domingo = gorjetas.query("dia == 'Domingo'").total_da_conta
```

```python
valor_conta_sabado = gorjetas.query("dia == 'Sábado'").total_da_conta
```

```python
r2 = ranksums(valor_conta_domingo, valor_conta_sabado)
```

```python
print("p-value = {}".format(r2.pvalue.round(3)))
```

**Como P-value > 0.05, então, aceita-se H<sup>null<sup/>:** A distribuição do valor da conta é igual no Sábado e no Domingo
